# The while statement
A while statement can be used to repeat a block of code while a certain condition is true.
<ol><li>Determine whether the condition is true or false.
<li>If false, exit the while statement and continue execution at the next statement.
<li>If the condition is true, run the body and then go back to step 1.

```python
def countdown(n):
    while n > 0:
        print(n)
        n = n - 1
    print('Blastoff!')
countdown(10)
```

```python
def print_even(n):
  while n > 0:
      print(n)
      n -= 2

print_even(100)
# Prints all even numbers from 0 - 100, starting from 100
```

## Infinite loop
The body of the loop should change the value of one or more variables so that the condition
becomes false eventually and the loop terminates. Otherwise the loop will repeat forever,
which is called an infinite loop.

```python
n=10
while n>0:
    print(n)
```

## break statement
Sometimes you don’t know it’s time to end a loop until you get half way through the body.
In that case you can use the break statement to jump out of the loop.

```python
while True:
    line = input(' ')
    if line == 'done':
        break
print(line)
```

### Square Root Example
one way of computing square roots is Newton’s method. Suppose that you
want to know the square root of a. If you start with almost any estimate, x, you can compute
a better estimate with the following formula:
y = (x + a/x) / 2

```python
a = 76 # the number for which we want to calculate the root
x = 9 # the initial estimate, can be a random number

while True:
    print(x)
    y = (x + a/x) / 2
    if y == x: # end the loop if there was no change in the estimated value compared to the previous round
        break
    x = y
```

# The range function
The range function generates a range of numbers between the beginning and end numbers. The beginning number is included while the end number is excluded.

```python
a = list(range(5, 10))
a
```

If the beginning number is not specified, the range starts at zero.

```python
a = list(range(5))
a
```

You can also use a third parameter as step size. Step size by default is 1.

```python
a = list(range(2,10,2))
a
```

As you can see, you can use the range function with different number of parameters. This is called <i>method overloading<i>.

# Continue
When the continue statement is executed, the program skips the rest of the statements in the block and starts the next round of iteration immediately.

```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

```python
for i in range(1,101):
  if i in range(40,51):
    continue # skips numbers 40 to 50
  print(i)
```

# Nested Loops
You can embed one loop inside another loop.

```python
for i in range(5):
    for j in range(8):
        print('*', end='')
    print()

# nested loops have uses in matrix algebra
```

# Class exercise
Write a function that calculates the sum of numbers 1 to n.

```python
def sum_num(n):
  result = 0
  for i in range(1,n+1,1):
    result+=i # result = result + i
  return result
```

```python
sum_num(5)
```
